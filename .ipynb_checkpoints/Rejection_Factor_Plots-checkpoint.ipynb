```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
##################################################
# set activitys
ActivityBi214 = 38250*1.260e-6 + 469*22.7e-6 + 10.6*578.5e-6
#print(ActivityBi214)
ActivityTl208 = 38250*0.435e-6 + 469*8.23e-6 + 10.6*85.9e-6
#print(ActivityTl208)
ActivityXe137 = 3.595e-06
ActivityRn222 = 9.270e-02
N_Xe137 = (1109e3/136)*6.022e23
Thalf   = 1e27
ActivityXe136 = np.log(2)*N_Xe137/Thalf/3.16e7
ActivityXe136
##################################################
# this is the rejection factors
##################################################
Xe137 = np.array([19000000, 574893, 513330, 263018, 13024, 2796])/19000000
bb0nu = np.array([1205000, 1006584, 948694, 669112, 466833, 344416])/1205000
RP = np.array([1000000000, 833290, 508210, 29357, 2100, 465])
FC = np.array([1000000000, 743902, 503613, 28698, 2005, 501])
IS = np.array([9996000000, 746787, 664665, 41258, 2917, 685])
Bi214 = RP+FC+IS
Bi214 = Bi214/Bi214[0]
RP = np.array([900000000 , 5419139, 3416885, 30728, 2539, 296])
FC = np.array([900000000 , 5849217, 3418123, 26739, 2205, 255])
IS = np.array([1000000000,  780228,  568383, 12865,  933, 147])
Tl208 = RP+FC+IS
Tl208 = Tl208/Tl208[0]
Rn222 = np.array([998200000, 2575495, 65817, 54, 4, 3])/998200000
```

```python
##################################################
# this is just for the second plot axis
##################################################
Xhold = [0,1,2,3,4,5]
tt = np.array([2,4,6,8])
MAJ = [1e-4,1e-3,1e-2,1e-1,1e0,1e1,1e2,1e3,1e4,1e5]
MIN = []
for O in MAJ:
    MIN.append(tt*O)
MIN = [y for x in MIN for y in x]
```

```python
##################################################
# here is the plot 
plt.figure(figsize=(12,7))
fig, ax=plt.subplots(figsize=(12,7))
ax.set_yscale('log')
lab = r'$0 \nu\beta\beta$ [T$_{1/2}$=10$^{27}$yr]'
ax.errorbar(Xhold, bb0nu*ActivityXe136*1e6,fmt='*',ms=15,ls='-',lw=3,color='darkorchid',label=lab)
lab = r'$^{208}Tl$'
ax.errorbar(Xhold, Tl208*ActivityTl208*1e6,fmt='>',ms=9,ls='--',lw=2,color='darkblue',label=lab,alpha=0.8)
lab = r'$^{214}Bi$'
ax.errorbar(Xhold, Bi214*ActivityBi214*1e6,fmt='<',ms=9,ls='--',lw=2,color='darkred',label=lab,alpha=0.8)
lab = r'$^{137}Xe$'
ax.errorbar(Xhold, Xe137*ActivityXe137*1e6,fmt='^',ms=9,ls='--',lw=2,color='steelblue',label=lab)
lab = r'$^{222}Rn$'
ax.errorbar(Xhold, Rn222*ActivityRn222*1e6,fmt='s',ms=9,ls='--',lw=2,color='darkgreen',label=lab,alpha=0.8)
ax.set_ylabel(r'Total Activity [$\mu$Bq]',fontsize=18)
my_xticks = ['Total', 'Broad Energy\n2.4 to 2.5MeV','Fiducial\n20mm', 'Single\nTrack', 'Two\nBlobs', 'Energy ROI\n'+'0.7%FWHM']
plt.xticks(Xhold, my_xticks, fontsize=16, rotation=0)
plt.yticks(fontsize=18)
plt.legend(loc='upper right',fontsize=18)
ax.set_yticks(MAJ, minor=False)
ax.set_yticks(MIN, minor=True)
ax.set_ylim(1.5e-4,2e5)
ax2 = ax.twinx()
ax2.set_yscale('log')
fact  = 3.154e+7/(22*1109)
fact2 = 1000*22
ax2.set_ylim((1.5e-4/1e6)*fact*fact2 ,(2e5/1e6)*fact*fact2)
ax2.set_ylabel(r'Counts / Ton / ROI / Yr',fontsize=18)
ax2.set_yticks([1e-2,1e-1,1e0,1e1,1e2,1e3,1e4,1e5,1e6], minor=False)
plt.yticks(fontsize=15)
ax.grid(which='both',linestyle='--')
ax.tick_params(axis='x',direction='in', length=6, width=2)
ax.tick_params(axis='y',direction='in', length=6, width=2)
ax2.tick_params(axis='y',direction='in', length=8, width=2)
plt.tight_layout()
#plt.savefig("BackGround-activity.png",bbox_inches='tight',dpi=400)
#plt.savefig("BackGround-activity.pdf")
plt.show()
```

```python
Indices=np.array([1,2,3])
Names=(['Copper\n(38,250 kg)','Teflon\n(469 kg)','Kapton\n'
        r'(10.6m$^2$)'])
Masses=np.array([38250,469,10.6])
ActivitiesTl_N100=np.array([0.435,8.23,85.9])
ActivitiesBi_N100=np.array([1.26,22.7,578.5])
ActivitiesTl_HD=np.array([0.435,0.035,107.3])
ActivitiesBi_HD=np.array([1.26,5.0,37.32])
RejFactorsTl_Pure=np.array([1.47,2.83,3.29])*1e-7
RejFactorsBi_Pure=np.array([0.68,5.01,4.65])*1e-7
RejFactorsTl_LoDiff=np.array([0.50,0.86,0.95])*1e-7
RejFactorsBi_LoDiff=np.array([0.19,1.42,1.56])*1e-7
```

```python
fig = plt.figure(figsize=(14,7))
fig.subplots_adjust(wspace=.39)
ax1 = fig.add_subplot(121)

ax1.barh(Indices+0.1,Masses*ActivitiesTl_N100,height=0.2,color='DarkBlue',label=r'$^{208}$Tl activity',alpha=0.8,zorder=3)
ax1.barh(Indices-0.1,Masses*ActivitiesBi_N100,height=0.2,color='DarkRed',label=r'$^{214}$Bi activity',alpha=0.8,zorder=3)

ax1.barh(Indices+0.1-3.5,Masses*ActivitiesTl_HD,height=0.2,color='DarkBlue',alpha=0.8,zorder=3)
ax1.barh(Indices-0.1-3.5,Masses*ActivitiesBi_HD,height=0.2,color='DarkRed',alpha=0.8,zorder=3)

ax1.fill_between([1,1e5],[-0,-0],[0.5,0.5],color='lightgrey',zorder=2,alpha=0.5)
ax1.fill_between([1,1e5],[3.5,3.5],[4.,4.],color='lightgrey',zorder=2,alpha=0.5)
ax1.text(1.5,3.65,'NEXT-100-style Background Model',fontsize=15)
ax1.text(1.5,0.15,'NEXT-HD-style Background Model',fontsize=15)
ax1.plot([1,1e5],[0.5,0.5],color='black',zorder=4,linewidth=1)
ax1.plot([1,1e5],[0.,0.],color='black',zorder=4,linewidth=1)
ax1.plot([1,1e5],[3.5,3.5],color='black',zorder=4,linewidth=1)


ax1.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.,fontsize=15)

ax1.set_xlabel("Total Activity ($\mu$Bq) \n Before Cuts ",fontsize=18)
plt.yticks(np.concatenate([Indices-3.5,Indices]),np.concatenate([Names,Names]))
ax1.tick_params(labelsize=15)
ax1.semilogx()
ax1.set_ylim(-3,4)
ax1.set_xlim(1,9e4)
ax1.grid(which='both', axis='x')



ax2 = fig.add_subplot(122)

plt.barh(Indices+0.1,Masses*ActivitiesTl_N100*RejFactorsTl_Pure,height=0.2,color='DarkBlue',alpha=0.5,zorder=3)
plt.barh(Indices-0.1,Masses*ActivitiesBi_N100*RejFactorsBi_Pure,height=0.2,color='DarkRed',alpha=0.5,zorder=3)

plt.barh(Indices+0.1-3.5,Masses*ActivitiesTl_HD*RejFactorsTl_Pure,height=0.2,color='DarkBlue',alpha=0.5,zorder=3)
plt.barh(Indices-0.1-3.5,Masses*ActivitiesBi_HD*RejFactorsBi_Pure,height=0.2,color='DarkRed',alpha=0.5,zorder=3)

plt.barh(Indices+0.1,Masses*ActivitiesTl_N100*RejFactorsTl_LoDiff,height=0.2,color='DarkBlue',label=r'$^{208}$Tl activity',alpha=0.8,zorder=3)
plt.barh(Indices-0.1,Masses*ActivitiesBi_N100*RejFactorsBi_LoDiff,height=0.2,color='DarkRed',label=r'$^{214}$Bi activity',alpha=0.8,zorder=3)

plt.barh(Indices+0.1-3.5,Masses*ActivitiesTl_HD*RejFactorsTl_LoDiff,height=0.2,color='DarkBlue',alpha=0.8,zorder=3)
plt.barh(Indices-0.1-3.5,Masses*ActivitiesBi_HD*RejFactorsBi_LoDiff,height=0.2,color='DarkRed',alpha=0.8,zorder=3)


plt.fill_between([1e-7,1e-2],[-0,-0],[0.5,0.5],color='lightgrey',zorder=2,alpha=0.5)
plt.fill_between([1e-7,1e-2],[3.5,3.5],[4.,4.],color='lightgrey',zorder=2,alpha=0.5)
plt.text(1.5e-7,3.65,'NEXT-100-style Background Model',fontsize=15)
plt.text(1.5e-7,0.15,'NEXT-HD-style Background Model',fontsize=15)
plt.plot([1e-7,1e-2],[0.5,0.5],color='black',zorder=4,linewidth=1)
plt.plot([1e-7,1e-2],[0.,0.],color='black',zorder=4,linewidth=1)
plt.plot([1e-7,1e-2],[3.5,3.5],color='black',zorder=4,linewidth=1)


ax2.legend(bbox_to_anchor=(0., 1.02, 1., .102), loc='lower left',
           ncol=2, mode="expand", borderaxespad=0.,fontsize=15)

ax2.set_xlabel("Total Activity ($\mu$Bq) \n After Cuts (Pure and Low Diffusion) ",fontsize=18)
plt.yticks(np.concatenate([Indices-3.5,Indices]),np.concatenate([Names,Names]))
ax2.tick_params(labelsize=15)
ax2.semilogx()
ax2.set_ylim(-3,4)
ax2.set_xlim(1e-7,9e-3)
ax2.grid(which='both', axis='x')


plt.tight_layout()
#plt.savefig("Both_Activities_Plot.png",bbox_inches='tight',dpi=400)
#plt.savefig("Both_Activities_Plot.pdf")
plt.show()

```
